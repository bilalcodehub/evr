```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = "1"
```

```python
from fastai.data.all import *
import subprocess, os

def reduce_bitrate(input_folder, new_bitrate='1000k'):
    # Create the "mini" folder if it doesn't exist
    output_folder = input_folder.replace("orig", "mini")
    Path(output_folder).mkdir(parents=True, exist_ok=True)

    # Get a list of video files in the input folder and sort them by name
    video_files = sorted(get_files(input_folder, extensions=['.mpg','.mp4','.mov']))

    for video_file in video_files:
        input_video = video_file
        output_video = os.path.join(output_folder, video_file.stem + '.mp4')  # Change the file extension to .mp4
        
        # Check if the output file already exists and delete it
        if os.path.exists(output_video):
            os.remove(output_video)
        
        # Get the duration of the input video using ffprobe
        ffprobe_cmd = f'ffprobe -i {input_video} -show_entries stream=duration -v quiet -of csv="p=0"'
        duration = subprocess.check_output(ffprobe_cmd, shell=True, encoding='utf-8')
        duration = duration.strip()  # Remove any leading/trailing whitespace

        # Generate a silent audio track with the same duration as the input video
        ffmpeg_cmd = (
            f'ffmpeg -i {input_video} -f lavfi -t {duration} -i anullsrc=r=44100:cl=stereo -shortest '
            f'-c:v libx264 -b:v {new_bitrate} -c:a aac -strict experimental {output_video}'
        )
        subprocess.run(ffmpeg_cmd, shell=True)

# Example usage:
input_folder = 'dataset/videos/surg-012/orig/'
new_bitrate = '1000k'

reduce_bitrate(input_folder, new_bitrate)
```
